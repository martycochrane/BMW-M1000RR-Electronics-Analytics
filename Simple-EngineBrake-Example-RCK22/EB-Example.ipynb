```python
from operator import indexOf
import pandas as pd
import numpy as np
from turtle import width
import plotly.graph_objects as go

df = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='EB3', header=None)

#Pull Off First Row for Lean
targetLean = df.iloc[0]
targetLean = targetLean.drop(index=df.index[0])

#Pull Off First Column for Mu_x
target_OLP = df.iloc[1]
target_OLP = target_OLP.drop(index=df.index[0])

#Read Log file
log = pd.read_csv("EB-Example.csv")

#Remove second row that has value types and convert whole dataframe to number
log.drop(log.index[0], inplace=True)
log.drop(log.index[1], inplace=True)
log.drop(log.index[0], inplace=True)
log = log.apply(pd.to_numeric)

#Convert All Lean Angles to positive
log['phi_lean'] = log['phi_lean'].abs()

slip = log['slip']
lean = log['phi_lean']
realEngineBrake = log['fx_engbr']
olpEngineBrakeTarget = log['fx_engbr_olp']
clpEngagement = log['fx_engbr_clp']



fig = go.Figure(data=[
    go.Scatter(x=lean, y=realEngineBrake, name = 'Actual Engine Brake'),
    go.Scatter(x=lean, y=olpEngineBrakeTarget, name = 'Open Loop EB Target'),
    go.Scatter(x=lean, y=clpEngagement, name = 'Closed Loop Engagement'),
    go.Scatter(x=targetLean, y=target_OLP, name = 'Open Loop Target', line_color='Red'),
    ])


fig.update_layout(title='Engine Brake - Turn 1 Schleiz', autosize=False,
                  width=1000, height=1000,
                  margin=dict(l=65, r=50, b=65, t=90))


fig.show()

print(lean)
```
