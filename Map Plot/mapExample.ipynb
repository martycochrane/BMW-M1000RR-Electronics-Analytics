```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

log = pd.read_csv("mapData.csv", header=12)

log.drop(log.index[0], inplace=True)
log = log.apply(pd.to_numeric)

latGPS = log['GPS Latitude']
longGPS = log['GPS Longitude']
tps = log['tpd']

fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(longGPS, latGPS, c=tps)
plt.set_cmap('bwr')
plt.title('Circuit Map with TPS Trace')
plt.show()
```
