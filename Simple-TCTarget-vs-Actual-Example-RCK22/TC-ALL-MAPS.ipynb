```python
from operator import indexOf
import pandas as pd
import numpy as np
from turtle import width
import plotly.graph_objects as go

df1 = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='MAP1', header=None)
df2 = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='MAP2', header=None)
df3 = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='MAP3', header=None)
df4 = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='MAP4', header=None)
df5 = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='MAP5', header=None)
df6 = pd.read_excel("bmwConfig-ALL-MAPS.xlsx",sheet_name='MAP6', header=None)

#Pull Off First Column for Mu_x
Target_mu_x = df1[0]
Target_mu_x = Target_mu_x.drop(index=df1.index[0])

Target_mu_x2 = df2[0]
Target_mu_x2 = Target_mu_x2.drop(index=df2.index[0])

Target_mu_x3 = df3[0]
Target_mu_x3 = Target_mu_x3.drop(index=df3.index[0])

Target_mu_x4 = df4[0]
Target_mu_x4 = Target_mu_x4.drop(index=df4.index[0])

Target_mu_x5 = df5[0]
Target_mu_x5 = Target_mu_x5.drop(index=df5.index[0])

Target_mu_x6 = df6[0]
Target_mu_x6 = Target_mu_x6.drop(index=df6.index[0])

#Pull Off First Row for Lean
targetLean = df1.iloc[0]
targetLean = targetLean.drop(index=df2.index[0])

targetLean2 = df2.iloc[0]
targetLean2 = targetLean2.drop(index=df2.index[0])

targetLean3 = df3.iloc[0]
targetLean3 = targetLean3.drop(index=df3.index[0])

targetLean4 = df4.iloc[0]
targetLean4 = targetLean4.drop(index=df4.index[0])

targetLean5 = df5.iloc[0]
targetLean5 = targetLean5.drop(index=df5.index[0])

targetLean6 = df6.iloc[0]
targetLean6 = targetLean6.drop(index=df6.index[0])


#Create 2D Array of Slip Target Values
for x in range(1, len(targetLean)+1):
    yArray=[]
    for y in range(1,len(Target_mu_x)+1):
        yArray.append(df1.iloc[x,y])
        if len(yArray) == 10:
            if x == 1:
                targetSlip = np.array([yArray])
            else:
                targetSlip = np.append(targetSlip,[yArray], axis=0)

for x in range(1, len(targetLean2)+1):
    yArray=[]
    for y in range(1,len(Target_mu_x2)+1):
        yArray.append(df2.iloc[x,y])
        if len(yArray) == 10:
            if x == 1:
                targetSlip2 = np.array([yArray])
            else:
                targetSlip2 = np.append(targetSlip2,[yArray], axis=0)

for x in range(1, len(targetLean3)+1):
    yArray=[]
    for y in range(1,len(Target_mu_x3)+1):
        yArray.append(df3.iloc[x,y])
        if len(yArray) == 10:
            if x == 1:
                targetSlip3 = np.array([yArray])
            else:
                targetSlip3 = np.append(targetSlip3,[yArray], axis=0)

for x in range(1, len(targetLean4)+1):
    yArray=[]
    for y in range(1,len(Target_mu_x4)+1):
        yArray.append(df4.iloc[x,y])
        if len(yArray) == 10:
            if x == 1:
                targetSlip4 = np.array([yArray])
            else:
                targetSlip4 = np.append(targetSlip4,[yArray], axis=0)

for x in range(1, len(targetLean5)+1):
    yArray=[]
    for y in range(1,len(Target_mu_x5)+1):
        yArray.append(df5.iloc[x,y])
        if len(yArray) == 10:
            if x == 1:
                targetSlip5 = np.array([yArray])
            else:
                targetSlip5 = np.append(targetSlip5,[yArray], axis=0)

for x in range(1, len(targetLean6)+1):
    yArray=[]
    for y in range(1,len(Target_mu_x6)+1):
        yArray.append(df6.iloc[x,y])
        if len(yArray) == 10:
            if x == 1:
                targetSlip6 = np.array([yArray])
            else:
                targetSlip6 = np.append(targetSlip6,[yArray], axis=0)

#Create Mesh Grid from Target Values
targetLean, Target_mu_x = np.meshgrid(targetLean, Target_mu_x)
targetLean2, Target_mu_x2 = np.meshgrid(targetLean2, Target_mu_x2)
targetLean3, Target_mu_x3 = np.meshgrid(targetLean3, Target_mu_x3)
targetLean4, Target_mu_x4 = np.meshgrid(targetLean4, Target_mu_x4)
targetLean5, Target_mu_x5 = np.meshgrid(targetLean5, Target_mu_x5)
targetLean6, Target_mu_x6 = np.meshgrid(targetLean6, Target_mu_x6)

fig = go.Figure(data=[
    go.Surface(x=Target_mu_x, y=targetLean, z=targetSlip, showscale=False, showlegend=True, opacity=0.5,colorscale=[[0, 'rgb(0,0,255)'], [1, 'rgb(0,0,255)']], name='Map 1'),
    go.Surface(x=Target_mu_x2, y=targetLean2, z=targetSlip2, showscale=False, showlegend=True, opacity=0.5,colorscale=[[0, 'rgb(255,0,0)'], [1, 'rgb(255,0,0)']], name='Map 2'),
    go.Surface(x=Target_mu_x3, y=targetLean3, z=targetSlip3, showscale=False, showlegend=True, opacity=0.5,colorscale=[[0, 'rgb(0,255,0)'], [1, 'rgb(0,255,0)']], name='Map 3'),
    go.Surface(x=Target_mu_x4, y=targetLean4, z=targetSlip4, showscale=False, showlegend=True, opacity=0.5,colorscale=[[0, 'rgb(0,0,0)'], [1, 'rgb(0,0,0)']], name='Map 4'),
    go.Surface(x=Target_mu_x5, y=targetLean5, z=targetSlip5, showscale=False, showlegend=True, opacity=0.5,colorscale=[[0, 'rgb(255,165,0)'], [1, 'rgb(255,165,0)']], name='Map 5'),
    go.Surface(x=Target_mu_x6, y=targetLean6, z=targetSlip6, showscale=False, showlegend=True, opacity=0.5,colorscale=[[0, 'rgb(255,192,203)'], [1, 'rgb(255,192,203)']], name='Map 6'),
    ])
fig.update_layout(title='BMW All TC Maps', autosize=False,
                  width=1000, height=1000,
                  margin=dict(l=65, r=50, b=65, t=90),
                  )

fig.update_layout(scene = dict(
                    xaxis_title='mu_x',
                    yaxis_title='Lean Angle (degrees)',
                    zaxis_title='Slip (%)'))

#Needed to make the layout match what you see in the BMW Software
fig.update_layout(
    scene={
        'xaxis': {'autorange': 'reversed'}, # reverse automatically
    }
)


fig.show()
```
